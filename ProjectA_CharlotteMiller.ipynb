This dataset describes the mathematics performance of students from grades 3-8 in NYC schools. The graphs I created depict a variety of things, but keep demographic as a consistent variable as I found it to be the most interesting. I investigated how many students from the various demographics were tested according to year, and also compared the mean scale score amongst them. I could only do this where applicable, as some of the mean scale scores have a value of NaN, which forbids it from being used in graphing that requires a particular axis to be a numerical value.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import requests
```

**a) distplot**

```python
sns.distplot(math["number_tested"])
```

**b) boxplot**

```python
sns.boxplot(x = "year" , y = "grade" , hue = "demographic", data = math, color = "red")
```

**c) violin_plot**

```python
sns.violinplot(x = "year" , y = "demographic" , hue = "grade" , data = math,
              bw=1, scale = "count", scale_hue = False)
```

```python
sns.violinplot(x = "year" , y = "grade" , hue = "demographic" , data = math,
              bw=1, scale = "count", scale_hue = False)
```
